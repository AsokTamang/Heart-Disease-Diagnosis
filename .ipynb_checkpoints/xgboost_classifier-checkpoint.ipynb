```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
RANDOM_STATE=55
```

**Reading the csv file**

```python
df = pd.read_csv('data/heart.csv')
df.head()
```

```python
df.columns
print(f'The features in given training examples are: {df.columns}')
```

**These are the columns or features which are categorical but donot have the numerical values**

```python
categorical_variables = ['ChestPainType','Sex','RestingECG','ExerciseAngina','ST_Slope']
```

**Creating one_hot encoding for the categorical variables which do not have bianry variable representation**

```python
df = pd.get_dummies(data=df,prefix = categorical_variables, columns = categorical_variables)  #here we are using one_hot encoding on the categorical columns
df.head()
```

```python
features = [x for x in df.columns if x not in df.HeartDisease]
X = df[features]  #the training features
y = df['HeartDisease']  #the target variable
print('The first five training examples are:', X[:5])
print('The first five target variable is:', y[:5])
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8, random_state=55)
print(len(X_train))
```

```python
min_samples_split_list = [2,10, 30, 50, 100, 200, 300, 700] ## If the number is an integer, then it is the actual quantity of samples,
max_depth_list = [1,2, 3, 4, 8, 16, 32, 64, None] # None means that there is no depth limit.
```

**Training the model based on various minimum sample split numbers**

```python
accuracy_list_train = []
accuracy_list_val = []
for min_samples_split in min_samples_split_list:
    model = DecisionTreeClassifier(min_samples_split = min_samples_split,
                                   random_state = RANDOM_STATE).fit(X_train,y_train) 
    predictions_train = model.predict(X_train) ## The predicted values for the train dataset
    predictions_val = model.predict(X_test) ## The predicted values for the test dataset
    accuracy_train = accuracy_score(predictions_train,y_train)  #checking how accurate our model is
    accuracy_val = accuracy_score(predictions_val,y_test)
    accuracy_list_train.append(accuracy_train)
    accuracy_list_val.append(accuracy_val)
```

**Plot of the training as well as test accuracy based on various minimum samples split list**

```python
plt.title('Train x Validation metrics')
plt.xlabel('min_samples_split')
plt.ylabel('accuracy')
plt.xticks(ticks = range(len(min_samples_split_list )),labels=min_samples_split_list)
plt.plot(accuracy_list_train)
plt.plot(accuracy_list_val)
plt.legend(['Train','Validation'])
```

**Training the model on various number of maximum depth of the decision tree**

```python
accuracy_list_train = []
accuracy_list_val = []
for max_depth in max_depth_list:
    model = DecisionTreeClassifier(max_depth = max_depth,
                                   random_state = RANDOM_STATE).fit(X_train,y_train) 
    predictions_train = model.predict(X_train) ## The predicted values for the train dataset
    predictions_val = model.predict(X_test) ## The predicted values for the test dataset
    accuracy_train = accuracy_score(predictions_train,y_train)
    accuracy_val = accuracy_score(predictions_val,y_test)
    accuracy_list_train.append(accuracy_train)
    accuracy_list_val.append(accuracy_val)
```

**Plot of the training as well as test accuracy based on various maximum depth of a decision tree**

```python
plt.title('Train x Validation metrics')
plt.xlabel('maximum depth')
plt.ylabel('accuracy')
plt.xticks(ticks = range(len(max_depth_list )),labels=max_depth_list)
plt.plot(accuracy_list_train)
plt.plot(accuracy_list_val)
plt.legend(['Train','Validation'])
```

```python
decision_tree_model =  DecisionTreeClassifier(min_samples_split = 25,max_depth = 4,random_state = 55).fit(X_train,y_train)
train_prediction = decision_tree_model.predict(X_train)
test_prediction = decision_tree_model.predict(X_test)
print(f'The accuracy of our final model on training dataset is: {accuracy_score(train_prediction,y_train)}')
print(f'The accuracy of our final model on test dataset is: {accuracy_score(test_prediction,y_test)}')
```

**RandomForestClassifier**

```python
accuracy_list_train = []
accuracy_list_val = []
for min_samples_split in min_samples_split_list:
    model = RandomForestClassifier(min_samples_split = min_samples_split,
                                   random_state = RANDOM_STATE).fit(X_train,y_train) 
    predictions_train = model.predict(X_train) ## The predicted values for the train dataset
    predictions_val = model.predict(X_test) ## The predicted values for the test dataset
    accuracy_train = accuracy_score(predictions_train,y_train)
    accuracy_val = accuracy_score(predictions_val,y_test)
    accuracy_list_train.append(accuracy_train)
    accuracy_list_val.append(accuracy_val)
```

**Plot of model accuracy on training as well as test datasets based on different minimum sample split numbers**

```python
plt.title('Train x Validation metrics')
plt.xlabel('min_samples_split_list')
plt.ylabel('accuracy')
plt.xticks(ticks = range(len(min_samples_split_list )),labels=min_samples_split_list)
plt.plot(accuracy_list_train)
plt.plot(accuracy_list_val)
plt.legend(['Train','Validation'])
```

```python
accuracy_list_train = []
accuracy_list_val = []
for max_depth in max_depth_list:
    model = RandomForestClassifier(max_depth = max_depth,
                                   random_state = RANDOM_STATE).fit(X_train,y_train) 
    predictions_train = model.predict(X_train) ## The predicted values for the train dataset
    predictions_val = model.predict(X_test) ## The predicted values for the test dataset
    accuracy_train = accuracy_score(predictions_train,y_train)
    accuracy_val = accuracy_score(predictions_val,y_test)
    accuracy_list_train.append(accuracy_train)
    accuracy_list_val.append(accuracy_val)
```

**Plot of model accuracy on training as well as test datasets based on different maximum depth of the decision trees**

```python
plt.title('Train x Validation metrics')
plt.xlabel('max_depth_list')
plt.ylabel('accuracy')
plt.xticks(ticks = range(len(max_depth_list )),labels=max_depth_list)
plt.plot(accuracy_list_train)
plt.plot(accuracy_list_val)
plt.legend(['Train','Validation'])
```

```python
n_estimators_list = [10,50,100,500] #number of decision trees for a random forest
accuracy_list_train = []
accuracy_list_val = []
for n_estimators in n_estimators_list:
    model = RandomForestClassifier(n_estimators = n_estimators,
                                   random_state = RANDOM_STATE).fit(X_train,y_train) 
    predictions_train = model.predict(X_train) ## The predicted values for the train dataset
    predictions_val = model.predict(X_test) ## The predicted values for the test dataset
    accuracy_train = accuracy_score(predictions_train,y_train)
    accuracy_val = accuracy_score(predictions_val,y_test)
    accuracy_list_train.append(accuracy_train)
    accuracy_list_val.append(accuracy_val)
```

**Plot of model accuracy on training as well as test datasets based on different numbers of decision trees in random forest**

```python
plt.title('Train x Validation metrics')
plt.xlabel('n_estimators_list')
plt.ylabel('accuracy')
plt.xticks(ticks = range(len(n_estimators_list )),labels=n_estimators_list)
plt.plot(accuracy_list_train)
plt.plot(accuracy_list_val)
plt.legend(['Train','Validation'])
```

**Using grid search to find the best model for randomforest classification**

```python
param_grid = {
    'n_estimators' : [10,50,100,500],
    'min_samples_split' : [2,10, 30, 50, 100, 200, 300, 700],
    'max_depth' : [1,2, 3, 4, 8, 16, 32, 64, None] 
}
grid_search = GridSearchCV(RandomForestClassifier(), param_grid,cv = 5,refit=True) #creating our random forest classifier model
grid_search.fit(X_train,y_train)
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
print(f'The best parameter is: {best_params}')
```

**Prediction of both training as well test datasets using best model**

```python
train_predicted = best_model.predict(X_train)
test_predicted = best_model.predict(X_test)
print(f'The accuracy of the model on training dataset is: {accuracy_score(train_predicted,y_train)}')
print(f'The accuracy of the model on test dataset is: {accuracy_score(test_predicted,y_test)}')
```

**XGBOOST classifier**

```python
X_train, X_, y_train, y_ = train_test_split(X, y, test_size=0.4)
X_val, X_test, y_val, y_test = train_test_split(X_, y_, test_size=0.5)
del X_,y_
xgboost_model = XGBClassifier(n_estimators=500,learning_rate=0.1,verbosity=1,random_state = 55,early_stopping_rounds = 10)
xgboost_model.fit(X_train,y_train,eval_set=[(X_val,y_val)]) 
training_prediction = xgboost_model.predict(X_train)
test_prediction = xgboost_model.predict(X_test)
print(f'The accuracy of xgboost model on training dataset is: {accuracy_score(training_prediction,y_train)}')
print(f'The accuracy of xgboost model on test dataset is: {accuracy_score(test_prediction,y_test)}')
```
